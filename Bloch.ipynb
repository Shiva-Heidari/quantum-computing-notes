```python
# Bloch Sphere Representation
# Hadamard Gate (Single qubit gate)
%matplotlib inline
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
import matplotlib.pyplot as plt

qc = QuantumCircuit(2)
qc.h(0)  # Superposition

state = Statevector.from_instruction(qc)
plot_bloch_multivector(state)
plt.show()
```

```python
# Creating Entanglement
# In Entanglement we can't identify one qubit by itself that's why we don't see any vector here!
qc = QuantumCircuit(2)
qc.h(0)
qc.cx(0, 1) # CNOT (multi-qubit gate) 
# cx(a, b)	Flip qubit b if qubit a is 1 
# cx(0, 1)	Qubit 0 controls Qubit 1
state = Statevector.from_instruction(qc)
plot_bloch_multivector(state)
plt.show()
print(state)
```

```python
# Enable inline plotting
%matplotlib inline

from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector
import matplotlib.pyplot as plt

# Create a simple circuit with 1 qubit and 1 classical bit
qc = QuantumCircuit(1, 1)

# Get the statevector (qubit is in default |0⟩ state)
stateVectorResult = Statevector.from_instruction(qc)

# Plot the state vector on the Bloch sphere
plot_bloch_multivector(stateVectorResult)
plt.show()
```

```python
qc = QuantumCircuit(1)
qc.x(0) #Rotate qubit from 0 to 1 using the x (NOT) gate
qc.h(0)  # Superposition or Hadamard gate

state = Statevector.from_instruction(qc)
plot_bloch_multivector(state)
plt.show()
```
